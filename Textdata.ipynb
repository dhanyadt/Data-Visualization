```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
text_data="Python is a multi-paradigm programming language and it is a object-oriented programming and structured programming are fully supported"
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('Word Cloud')
plt.show()
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(), word_counts.values())
plt.title('Histogram of Word Frequencies')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
