```python
x = df["release_year"].mode()[0]
df["release_year"].fillna(x, inplace=True)
unwanted_characters = ['-', '(', ')', 'I','II']

# Remove unwanted characters from the "release_year" column
for character in unwanted_characters:
    df['release_year'] = df['release_year'].replace(character, '')

print(df.to_markdown())


x = df["gross"].mode()[0]
df["gross"].fillna(x, inplace=True)
unwanted_characters = ['$','M']

# Remove unwanted characters from the "gross" column
for character in unwanted_characters:
    df['gross'] = df['gross'].replace(character, '')

print(df.to_markdown())
```
