```python
import requests
Link="https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1"
Link_text=requests.get(Link).text
print(Link_text)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(Link_text,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
def getdata(url):
    r = requests.get(url)
    return r.text

htmldata = getdata("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item['src'])
```
